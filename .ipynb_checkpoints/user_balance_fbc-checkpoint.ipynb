```python
#常用数据存放目录
catelog='./data/'
#临时数据存放目录
tempData='./temp/'
#数据库数据数据定义
Database='lingoacedb'

def data_to_csv(df, name):
    
    catelog='./data/'

    name_suffix = datetime.now().strftime('%Y%m%d')
    
    df.to_csv('%s%s%s.csv' % (catelog, name, name_suffix), encoding='utf_8_sig')
```

```python
all_parent_sql = '''select 
                        id, 
                        username, 
                        create_time,
                        balance+bonus_balance as balance, 
                        sg_balance,
                        tmp_fbc.last_fbc_create_time,
                        tmp_fbc.fbc_amount
                    from user_parent_info upi
                    left join (
                    select max(fbc.create_time) as last_fbc_create_time, sum(fbc.amount) as fbc_amount, fbc.parent_user_id from finance_balance_change fbc
                    group by fbc.parent_user_id
                    ) tmp_fbc on tmp_fbc.parent_user_id=upi.id
                    where upi.is_staff<>1 and upi.status=1'''
```

```python
all_parent= data_out_sql(all_parent_sql)
all_parent.head()
```

```python
error_6_sql = '''select upi.id, sum(fbc.amount) as error_amount from finance_balance_change fbc 
                    left join user_parent_info upi on fbc.parent_user_id=upi.id
                    where fbc.create_time < upi.create_time
                    group by upi.id'''
error_6 = data_out_sql(error_6_sql)

error_6_flag = pd.merge(all_parent, error_6, how='left', on='id')
error_6_flag.head()
```

```python
repeat_balance_sql = '''select tmp.parent_user_id, sum(tmp.sum_amount) as repeat_amount from (
                            SELECT
                                count(fbc.id),
                                max( fbc.id ),
                                min( fbc.id ),
                                (count(fbc.id)-1)*fbc.amount as sum_amount,
                                fbc.role,
                                fbc.reason,
                                fbc.create_time,
                                fbc.update_time,
                                fbc.parent_user_id,
                                fbc.reference,
                                fbc.amount
                            FROM
                                    finance_balance_change fbc 
                            WHERE
                                    fbc.role IN ( 1, 2 ) and fbc.update_time is null
                            GROUP BY fbc.create_time,fbc.update_time,fbc.adviser_user_id,fbc.xg_user_id,fbc.amount,fbc.reason,fbc.parent_user_id,fbc.reference,fbc.role 
                            HAVING
                                    count( fbc.id ) > 1) tmp group by tmp.parent_user_id'''

repeat_balance= data_out_sql(repeat_balance_sql)
repeat_balance.head()
```

```python
# 添加是否重复标签
repeat_flag = pd.merge(error_6_flag, 
                      repeat_balance, 
                      how='left', 
                      left_on='id', 
                      right_on='parent_user_id')

del repeat_flag['parent_user_id']
repeat_flag.head()
```

```python
# data_to_csv(all_parent_repeat_flag, 'all_parent_repeat_flag')
# 用户在老数据库的数据跟在新库里的数据
migrate_data_compare_sql = '''select 
                                au.username, 
                                sum(abc.amount) as old_abc_amount,
                                (select sum(fbc.amount) from finance_balance_change fbc left join user_parent_info upi on upi.id=fbc.parent_user_id where upi.username=au.username and fbc.create_time<=max(abc.created_on)) as old_fbc_amount
                            from pplingodb.accounts_accountbalancechange abc 
                            left join pplingodb.auth_user au on au.id=abc.user_id
                            group by au.id'''

migrate_data_compare = data_out_sql(migrate_data_compare_sql)
migrate_data_compare.head()
```

```python
# 丢失的数据
lose_data_flag = pd.merge(repeat_flag, 
                          migrate_data_compare, 
                          how='left', 
                          left_on='username', 
                          right_on='username')
lose_data_flag.head()
```

```python
data_to_csv(lose_data_flag, 'lose_data_flag')
```

```python
# balance合并， fbc未修改的用户
balance_merge_sql = '''select 
                                tmp_supi.parent_user_id, 
                                sum(tmp_supi.supi_fbc_amount) as supi_fbc_amount, 
                                sum(tmp_supi.supi_balance) as supi_balance,
                                sum(tmp_supi.sg_balance) as supi_sg_balance,
                                sum(tmp_supi.abc_sum_amount) as supi_abc_amount,
                                sum(tmp_supi.ppabc_fbc_amount) as supi_old_fbc_amount,
                                sum(tmp_supi.repeat_amount) as supi_repeat_amount
                        from (
                        select 
                                upi.id as parent_user_id, 
                                usi.student_parent_user_id as student_parent_user_id,
                                (select sum(fbc.amount) from finance_balance_change fbc where fbc.parent_user_id=student_parent_user_id) as supi_fbc_amount,
                                (supi.balance+supi.bonus_balance) as supi_balance,
                                supi.sg_balance,
                                ppabc.abc_sum_amount,
                                ppabc.fbc_amount as ppabc_fbc_amount,
                                tmp_repeat.repeat_amount
                        from user_parent_info upi
                        left join user_student_info usi on usi.parent_user_id=upi.id
                        left join user_parent_info supi on usi.student_parent_user_id=supi.id
                        left join (
                        select tmp.parent_user_id, sum(tmp.sum_amount) as repeat_amount from (
                                                    SELECT
                                                                                        fbc.parent_user_id,
                                                        (count(fbc.id)-1)*fbc.amount as sum_amount
                                                    FROM
                                                            finance_balance_change fbc 
                                                    WHERE
                                                            fbc.role IN ( 1, 2 ) and fbc.update_time is null
                                                    GROUP BY fbc.create_time,fbc.update_time,fbc.adviser_user_id,fbc.xg_user_id,fbc.amount,fbc.reason,fbc.parent_user_id,fbc.reference,fbc.role 
                                                    HAVING
                                                            count( fbc.id ) > 1) tmp group by tmp.parent_user_id) tmp_repeat on tmp_repeat.parent_user_id=supi.id
                        left join (select 
                                au.username, 
                                sum(abc.amount) as abc_sum_amount,
                                (select sum(fbc.amount) from finance_balance_change fbc left join user_parent_info upi on upi.id=fbc.parent_user_id where upi.username=au.username and fbc.create_time<=max(abc.created_on)) as fbc_amount,
                                au.is_staff
                        from pplingodb.accounts_accountbalancechange abc 
                        left join pplingodb.auth_user au on au.id=abc.user_id
                        group by au.id) ppabc on ppabc.username=supi.username
                        where  usi.student_parent_user_id is not null and usi.student_parent_user_id<>usi.parent_user_id
                        ) tmp_supi
                        group by tmp_supi.parent_user_id
                        '''
balance_merge = data_out_sql(balance_merge_sql)

balance_merge.head()
```

```python
balance_merge_flag = pd.merge(lose_data_flag, 
                              balance_merge, 
                              how='left',
                              left_on='id',
                              right_on='parent_user_id')
del balance_merge_flag['parent_user_id']
balance_merge_flag.head()
```

```python
data_to_csv(balance_merge_flag, 'balance_merge_flag')
```
